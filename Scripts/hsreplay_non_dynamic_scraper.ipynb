```python
class DeckAnalyzer:
    '''
    Analyze a certain deck from the hsreplay website
    '''
    def __init__(self, link):
        headers = {'User-Agent': 'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/81.0.4044.138 Safari/537.36'}
        r = requests.get(link, headers = headers)
        self.soup = BeautifulSoup(r.text, 'html.parser')


    def get_deck_info(self):
        '''
        Print various information about the deck
        '''
        deck_info = self.soup.find('div', id = 'deck-info').attrs
        return deck_info
    
    def get_deck_attrs(self, attrs):
        '''
        Define a list of desired attributes and get the information about these attributes
        
        Choose the desired attributes from
            'id', 'data-deck-id', 'data-deck-name', 'data-deck-format', 'data-hero-id', 'data-deck-cards',
            'data-deck-cards', 'data-archetype-id', 'data-archetype-name'
        '''
        deck_info = self.get_deck_info()
        for a in attrs:
            info = deck_info.get(a)
            print(f'{a} -> {info}')
        return None
    
    def get_card_names(self):
        '''
        Get a list of cards in the deck
        '''
        card_ids = self.get_deck_info().get('data-deck-cards')
        for c in self.soup.findAll('figcaption'):
        #, class_ = 'card-name')
            print(c)
        return self.soup
        
```
